```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample customer dataset
data = {
    "Customer": ["A", "B", "C", "D", "E", "F", "G", "H"],
    "Age": [22, 35, 45, 30, 50, 29, 41, 38],
    "MonthlyCharges": [500, 700, 1200, 400, 1500, 650, 1000, 850],
    "Tenure": [2, 5, 1, 7, 1, 6, 2, 8],
    "Churn": ["Yes", "No", "Yes", "No", "Yes", "No", "Yes", "No"]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Customer Dataset:\n")
print(df)

# Count churn customers
print("\nChurn Count:")
print(df["Churn"].value_counts())

# Plot churn analysis
df["Churn"].value_counts().plot(kind='bar')

plt.title("Customer Churn Analysis")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")

plt.show()
```
